```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
```

Entrenando con cajero analizando cajero 2.

```python
model_id = 'forecast_model_for_cluster_0_TimesNet_ATM_ftMS_sl60_pl1_dm64_el2_df64_ebfixed_0'
path_root = f'./results/{model_id}/'
file_name = 'pred.npy'

pred = np.load(f'{path_root}'+'pred.npy')

trues = np.load(f'{path_root}'+'true.npy')

pred = pred.reshape(-1, 1)
trues = trues.reshape(-1, 1)
```

```python
#Numero de cajero en el cluster.
k= 23

plt.figure(figsize=(12, 6))
plt.plot(trues[37*k:37*(k+1)], label='True Values', color='blue')
plt.plot(pred[37*k:37*(k+1)], label='Predictions', linestyle='--', color='orange')
plt.title('Predictions vs True Values')
plt.xlabel('Time Steps')
plt.ylabel('Values')
plt.legend()
plt.grid()
plt.show()

#Calcular el error absoluto
errors = np.sum(pred - trues)
print(f'Error absoluto: {errors}')
```
