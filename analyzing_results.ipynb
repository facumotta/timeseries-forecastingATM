```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
```

Entrenando con cajero analizando cajero 2.

```python
model_id = 'forecast_TimesNet_cajero_2_TimesNet_ATM_ftMS_sl30_pl1_dm512_el2_df256_ebfixed_0'
path_root = f'./results/{model_id}/'
file_name = 'pred.npy'
full_path = f'{path_root}{file_name}'


pred = np.load(full_path)

pred = pred.reshape(-1, 1)
pred
```

```python
file_name = 'pred.npy'
full_path2 = f'{path_root}{file_name}'


trues = np.load(full_path2)

trues = trues.reshape(-1, 1)

trues.shape, pred.shape
```

```python
plt.figure(figsize=(12, 6))
plt.plot(trues, label='True Values', color='blue')
plt.plot(pred, label='Predictions', linestyle='--', color='orange')
plt.title('Predictions vs True Values')
plt.xlabel('Time Steps')
plt.ylabel('Values')
plt.legend()
plt.grid()
plt.show()
```

Analizo cajero 3 habiendo entrenado con cajero 2.

```python
model_id = 'forecast_TimesNet_cajero_2_TimesNet_ATM_ftMS_sl30_pl1_dm512_el2_df256_ebfixed_0'
path_root = f'./results/{model_id}/'
file_name = 'pred.npy'
full_path = f'{path_root}{file_name}'


pred = np.load(full_path)

pred = pred.reshape(-1, 1)

file_name2 = 'pred.npy'
full_path2 = f'{path_root}{file_name2}'


trues = np.load(full_path2)

trues = trues.reshape(-1, 1)

trues.shape, pred.shape
```

```python
df = pd.read_csv(f'data_processed/df_resultado_cajero_3.csv')


real_real = df['TARGET'].values
real_real = real_real.reshape(-1, 1)

#Quiero los ultimos 37
real_real = real_real[-37:]

real_real.shape, trues.shape, pred.shape
```

```python
plt.figure(figsize=(12, 6))
plt.plot(trues, label='True Values', color='blue')
plt.plot(pred, label='Predictions', linestyle='--', color='orange')
plt.plot(real_real, label='Real Values', linestyle=':', color='green')
plt.title('Predictions vs True Values')
plt.xlabel('Time Steps')
plt.ylabel('Values')
plt.legend()
plt.grid()
plt.show()
```
